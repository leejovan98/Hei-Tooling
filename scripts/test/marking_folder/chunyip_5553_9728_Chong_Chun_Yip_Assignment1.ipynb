- Do refer to Exercise 5r of Lesson 1 if you need a refresher on the apply function.

```python
## Write your code below

def text_to_binary(loan_status_str):
    
    if loan_status_str == "Y":
        return 1
    
    else:
        return 0

newlabel = loans_df["Loan_Status"].apply(text_to_binary)

loans_df["Loan_Status"] = newlabel
loans_df.head()
```
